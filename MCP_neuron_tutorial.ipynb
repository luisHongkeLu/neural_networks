## Introduction to McCulloch-Pitts Neuron
McCulloch-Pitts (MCP) neuron is named after two scientists: Warren McCulloch, a neurophysiologist, and Walter Pitts, a logician. They wanted to use mathematics to model an artificial neuron. Of course, they managed to do it, even though it's a very simple module. 


From the fields of these two scientists, we can see that MCP neuron is based on neuroscience and mathematics, and our tutorial will start from the neuroscience background information.

##How does the individual neuron work? 
1. The neuron will receive an input signal through the dendrite.
2.  Then, the input information will be processed by the soma (cell body).
3.  The information will pass through the axon to the synapses. 
4.  Then, the synapses will pass the information of whether it fires to the dendrite of the next neuron. 
*Here is worth mentioning the "all or none" aspect of bio neurons. The synapses only fire when the activation energy is over the threshold energy, and "fire" is the only feature that we can observe. 








##Dive into the Soma
### The Threshold Energy and fire 
The threshold is the criteria for the neuron to decide whether it fires or not. If the weighted total energy processed by the soma is higher than the threshold, the soma will determine that the neuron should fire and vice versa. 

There are two types of neurons; one passes the inhibitory signals, which can decrease the current energy, and another passes the excitatory signals, increasing the current energy. 

Picture from http://techlab.bu.edu/files/resources/software/epsp_ipsp.jpg


epsp_ipsp.jpg

###EPSP
EPSP (excitatory postsynaptic potentials) occurs when positive ions flow into the postsynaptic neuron and generates a temporary depolarization of postsynaptic membrane. The postsynaptic neuron receives an EPSP after an excitatory presynaptic neuron fires an action potential. The EPSP signal is delivered along down the dendrite and can be summed with other inputs. The EPSP signal will increase the neurons membrane potential, and an action potential will be produced when the membrane potential reaches the threshold.

###IPSP
IPSP (inhibitory postsynaptic potentials) occurs when negative ions flow into the postsynaptic neuron and generate a temporary hyperpolarization of the postsynaptic membrane. The postsynaptic neuron receives an IPSP after an inhibitory presynaptic cell fires an action potential. The IPSP signal is delivered along down the dendrite and will decrease the neurons membrane potential. The IPSP signal can be summed with other inputs.

###Synaptic integration
Synaptic integration is the process of determining neurons outputs from the synaptic inputs they receive, which is how neurons add up inputs before producing an action potential.

## Dot Product 

The dot product is a powerful tool that we can use while calculating the weighted sum of the activation energy. 

Dot product is the inner product of two vectors, which is defined as:
$$\vec{A} = (a, b)$$
$$\vec{B} = (c, d)$$
$$A \cdot B = ac + bd $$

```python
#Dot product in python
A = np.array([2,3])
B = np.array([4,5])
AdotB=A[0]*B[0] + A[1]*B[1]
print(f"the outcome of dot product based on definition : {AdotB}")
# And here is an easier way that based on numpy
AdotB_np = np.dot(A,B)
print(f"the outcome of dot product using numpy: {AdotB_np}")
print(f"We can see that the outcomes are the same")
```

## MCP Neural Networks in Math and Python

Here we have multiple inputs pass into the neuron, and the neuron's soma will produce an output based on the input energy. As discussed before, if the input energy is higher than the threshold energy, the neuron will produce an output of True; if the input energy is lower than the threshold energy, the neuron will produce an output of False. You can see the mathematical expression that expressed the same thing. 

IMG_6EE097DEC5DB-1.jpeg

For the graph above, if you change $\theta$ to 2, it will be an And gate!

```python
#example for Or(And) Gate
input = np.array([1,1])
threshold = 1 #you may change the threshold to 2, then the logic gate
              # will be an And Gate
if input[0]+input[1] > threshold: 
  output = 1
else:
  output = 0
print(output)
if output ==1: 
  print("The neuron is Firing")
else:
  print("The neuron is not Firing")
```

However, in some circumstances, one may have a preference. They may value more to the certain attribute of the inputs over another. As a result, we will introduce weights into the model above. 

IMG_0440.jpg

In the model above, the soma no longer adds the inputs; instead, it calculates the weighted input. And it will generate an output based on the weighted input. However, the underlying logic does not change. The soma is still valuing whether the weighted input is larger than the threshold. If weighted input > threshold, it will generate an output of True and vice versa. 

```python
# Example for MCP neuron that have weights plays around 
input = np.array([1,1])
weight = np.array([1,-1])
threshold = 0
if input[0]*weight[0]+input[1]*weight[1] >= threshold: 
  output = 1
else:
  output = 0
print(f"the weighted input is {input[0]*weight[0]+input[1]*weight[1]}")
print(f"The output is {output}")
if output ==1: 
  print("The neuron is Firing")
else:
  print("The neuron is not Firing")
```

```python
# Do you remember the dot product that we discussed before? 
# How to use that in the weighted input? 
# Check the answer of the weighted input that we calculated above with the dot
# product that we calculated in this question!

#Ans:
input = np.array([1,1])
weight = np.array([1,-1])
threshold = 0
if np.dot(input,weight) >= threshold: 
  output = 1
else:
  output = 0
print(f"the weighted input is {np.dot(input,weight)}")
print(f"The output is {output}")
if output ==1: 
  print("The neuron is Firing")
else:
  print("The neuron is not Firing")
```

Now we can work through an example to see how to make an MCP neuron using python!

The scientist believes that hawk will only hunt for the animal that both small and have no spike. All the animals we consider will only have two properties: body size and spike. For the [body size] part, I will assign True to the small body size animals and vice versa. For the [spike], I will assign True for porcupine and hedgehog. 

In this example, we consider four animals: lion, mouse, hedgehog, and porcupine. The array below follows [body_size, spike]:
lion = [0,1]
mouse= [1,1]
porcupine = [0,0]
hedgehog = [1,0]

```python
mouse= np.array([1,1])
lion = np.array([0,1])
hedgehog = np.array([1,0])
porcupine = np.array([0,0])
# Since the hawk will only catch the mouse, so it's an And Gate
# 1. using the np.vstack() function to combine these four arrays
prey = np.vstack((mouse,lion,hedgehog,porcupine))
# 2. remind that we are using the threshold! but not the logic gate 
# that from python!
threshold = 2
weights = np.array([1,1])
# 3. initialize the final output and middle output!
finalOutput = np.zeros(4)
weightedInput = np.zeros(4)
expected = np.array([1,0,0,0])
# 4. Write an logic gate by using threshold! Remember that we need to enumerate 
# through the array
for i in range(len(finalOutput)):
    weightedInput[i] = np.dot(prey[i,:],weights) # we are doing addition here
for i in range(len(finalOutput)):
    if weightedInput[i] >= threshold:
        finalOutput[i] = 1
print(f"The final output is {finalOutput}")
if (finalOutput == expected).all():
  print("You can see that we successfully identify that mouse is a good prey!")
else:
  print("Fail")
```

####Multilayers MCP neural network
Have you ever heard of this before? Maybe no, but it's important to know it right now. A person is made of a neural network that has infinite layers! Now let's do a multilayer MCP neural network!

Suppose we want to teach the computer about the magnet work. Given two magnets, and both of them have the clear south pole and north pole. How can the computer successfully identify when the two magnets will attract each other? Make the north pole as 1 and the south pole as 0. 

Here are four situations: 
northNorth: [1,1]
southNorth: [0,1]
northSouth: [1,0]
southSouth: [0,0]

The magnets will only attract each other when the poles are different. 

```python
#Have a try!
NN = np.array([1,1])
NS = np.array([1,0])
SN = np.array([0,1])
SS = np.array([0,0])
# 1. using the np.vstack() function to combine these four arrays
experiment = np.vstack((NN,NS,SN,SS))
# In order to keep it simple, I use the weights as [1,1]
weights = np.array([1,1])
not_weights = np.array([-1])
# remind from the tutorial above, we need to use three gates to make a XOR gate
threshold_and = 2
threshold_or = 1
threshold_not = 0
# initialize the results!
weightedInput = np.zeros(4)
weightedNotInput = np.zeros(4)
weightedInputXOR = np.zeros(4)
output_and = np.zeros(4)
output_notand = np.zeros(4)
output_or = np.zeros(4)
output_XOR = np.zeros(4)
expected = np.array([0,1,1,0])
#calculate the weighted input:
for i in range(len(experiment)):
  weightedInput[i] = np.dot(experiment[i,:],weights)
# Now apply the AND gate, which threshold = 2
output_and[weightedInput >= threshold_and] = 1
# remind that there is a Not gate! 
for i in range(len(output_and)): 
  weightedNotInput[i] = np.dot(output_and[i],not_weights)
  output_notand[weightedNotInput >= threshold_not] = 1
# here we get the not and gate! Then we should do a or gate! 
# You can choose to print out the variable anytime you want!
output_or[weightedInput >= threshold_or] = 1
# In the end, use And gate to connect the answer from not and gate and or gate!
# Here we add the two gates!
for i in range(len(output_or)):
  #here we are using the weight of 1 for both output from notand and or
  weightedInputXOR[i] = output_notand[i]*1 + output_or[i]*1
# At last use an And gate!
output_XOR[weightedInputXOR >= threshold_and] = 1
print(f"The final output is {output_XOR}")
if (output_XOR == expected).all():
  print("Victory! XOR Gate")
else:
  print("Fail")
```

##Limitations of MCP Neural Network

Back to the very beginning of the MCP neuron, can you think about some important points which MCP neural network includes? 

1. Inhibitory and excitatory signals are the only two weights that the MCP neural network can use. What if we want any other weight? Think about how to distinguish a cat from a dog. Most cats do "meow," so we may want to let the character of "meow" be one of the dominant characters. As a result, we may change the weight of "meow" to be much bigger than 1. However, the MCP neural network does not support us to do that. 
2. Can you think of how we recognize the cat? Are we using pure logic, or do we feel what a cat is? The cognition of human beings includes many hidden variables that we cannot find directly through direct observations. 
3. One of the most important shortcomings of MCP neurons is we need to set the threshold by hand. However, from the reading, we can see that the weighted networks solve this shortcoming by changing the threshold into one of the inputs.
